```python
# create a lastday variable in the datetime type feature
reference_date = max(df_merged["TransactionDate"].max(),
                     df_merged["InteractionDate"].max(),
                     df_merged["LastLoginDate"].max())

# create a recency feature
df_merged["day_since_transaction"] = (reference_date - df_merged["TransactionDate"]).dt.days
df_merged["day_since_interaction"] = (reference_date - df_merged["InteractionDate"]).dt.days
df_merged["day_since_lastlogin"] = (reference_date - df_merged["LastLoginDate"]).dt.days

# remove native datetime feature
df_merged = df_merged.drop(columns=["TransactionDate", "InteractionDate", "LastLoginDate"])
```

```python
df_merged.head(15)
```

Transforming dates into numeric metrics (day_since_transaction, day_since_interaction, day_since_lastlogin) makes the data more meaningful for churn analysis. From the preview, it can be seen that there are customers who have not had a transaction or interaction for a very long time (hundreds of days), while the login distance varies more. This illustrates that activity recency can be a strong indicator of churn, although it is not enough to just one activity alone—a combination of transactions, logins, and interactions is more relevant to explain customer behavior.

It is possible that the relationship with churn is non-linear or arises from a combination of several variables at once, for example customers who rarely log in and who have not made transactions for a long time. In cases like this, the correlation approach alone is insufficient to understand the pattern. Therefore, further analysis should use predictive models such as Random Forest or XGBoost to measure feature importance and explore interactions between variables. That way, we can find out which factors really contribute to churn prediction even if they are not obvious in a simple correlation analysis.
