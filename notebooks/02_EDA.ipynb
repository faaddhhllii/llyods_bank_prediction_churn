# EDA (Exploratory Data Analysis)

```python
# Distribution feature gender
sns.countplot(x="Gender", data=df_merged)
plt.title("Distribution Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Gender distribution shows that there are more female customers than male customers, so marketing strategies can be focused more on the female segment. Even so, the number of male customers also remains significant, so campaigns that are inclusive for both are still important so as not to lose market potential.

```python
# Distribution feature age
sns.histplot(x="Age", data=df_merged)
plt.title("Distribution Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

In terms of age, customers are evenly distributed between 18 and 69 years, with a predominance in the young age group (18–22 years) and those approaching retirement (50 years and over). This opens up opportunities for strategic differentiation: younger groups can be approached with digital-based promotions and affordable prices, while older age groups can be focused on convenience and quality of service. Meanwhile, the relatively smaller middle age group (30–45 years) could actually be a target for new acquisitions with a more relevant approach, such as family programs or additional benefits for professionals.

```python
# labelling group age
def label_age(Age):
  if 18 <= Age <= 29:
    return "Young Adults"
  elif 30 <= Age <= 40:
    return "Working Adults"
  else:
    return "Older Adults"

df_labeled = df_merged.copy()
df_labeled["Age_Group"] = df_labeled["Age"].apply(label_age)

# visualisasi age group
sns.countplot(x="Age_Group", data=df_labeled)
plt.title("Distribution Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

Based on age grouping, the majority of customers are in the Older Adults category (≥ 41 years) with almost 3,000 people. This shows that the majority of service users come from the older adult age group.

Meanwhile, the Young Adults category (18–29 years) is in second place with around 1,300 people, which shows that there is a young market segment that is quite large and has the potential to be targeted through digital promotions or trend-based marketing strategies.

The Working Adults category (30–40 years) is actually the group with the smallest number, around 1,000 people. This could be a business opportunity to increase penetration in this segment, for example by offering products or services that are relevant to young professionals and young families.

```python
# visualization of income distribution in each age group
age_groups = df_labeled["Age_Group"].unique()

for group in age_groups:
    group_data = df_labeled[df_labeled["Age_Group"] == group]
    income_counts = group_data["IncomeLevel"].value_counts()

    plt.figure()
    income_counts.plot.pie(autopct="%1.1f%%", startangle=90, shadow=True, colormap="Set3")
    plt.title(f"Income Composition in {group}")
    plt.ylabel("")
    plt.tight_layout()
    plt.show()
```

If we look at the composition of income in each age group, we can see quite different patterns. For Older Adults, the majority are in the Medium Income class with the largest proportion, namely 36.6%. This indicates that the older age group tends to have financial stability with a middle income.

For Young Adults, the largest distribution is in the Low Income class at 36.6%. This finding is quite natural considering that young age groups are usually still in the early stages of their careers, so their purchasing power is relatively lower than other groups.

Meanwhile, for Working Adults, the majority are also in the Medium Income class with the largest percentage being 37.1%. This shows that the productive age group tends to have middle income, in accordance with their developing career phase.

```python
# visualisasi distribusi fitur marital status
sns.countplot(x="MaritalStatus", data=df_merged)
plt.title("Distribution Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.show()
```

The distribution of marital status shows that customers are spread quite evenly across all categories, with dominance in the Widowed and Divorced groups, each reaching around 1,400 people. The Single group numbers around 1,200 people, while the fewest customers with Married status are around 1,150 people. These findings illustrate that services or products are used by various levels of marital status, but attract more interest from individuals who are not in an active marital relationship. In terms of business strategy, the Widowed and Divorced segments can be focused on campaigns that emphasize comfort and quality of life, the Single segment can be directed to products and promotions based on trends and modern lifestyles, while the Married segment can potentially be targeted with family products, couple bundling, or cross-selling programs.

```python
marital_status = df_merged["MaritalStatus"].unique()

for status in marital_status:
    group_data = df_merged[df_merged["MaritalStatus"] == status]
    income_counts = group_data["IncomeLevel"].value_counts()

    plt.figure()
    income_counts.plot.pie(
        autopct="%1.1f%%",
        startangle=90,
        shadow=True,
        colormap="Set3"
    )
    plt.title(f"Income Composition in {status}")
    plt.ylabel(" ")
    plt.tight_layout()
    plt.show()
```

Analysis of income composition based on marital status shows varying patterns. Customers with Single status are dominated by the High Income group (35.2%), while Married and Widowed customers are more likely to be in the Medium Income category at 34.2% and 35.4% respectively. Meanwhile, customers with Divorced status tend to be more dominant in the Low Income class (36.4%). Even though there is dominance in certain segments, in general the income distribution in these four groups is relatively even in the High, Medium and Low classes. These findings provide an illustration that business strategies can be specifically adjusted: the Single segment has the potential to be targeted with premium and exclusive products, the Married and Widowed segments are suitable for focusing on mid-range products with loyalty or bundling programs, while the Divorced segment is more relevant with an affordable price strategy through discounts, economical packages or easy installments.
